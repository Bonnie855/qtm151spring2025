```python
import matplotlib.pyplot as plt
import pandas as pd

print('Hello World')
```

```python
carfeatures = pd.read_csv('data/features.csv')
carfeatures # show all data in carfeatures
```

```python
carfeatures['cylinders'] # only show 'cylinders' column
```

```python
# crosstab: create a table that counts occurrences of values in a column
# index: group data by the unique values in the cylinders column
# columns: add customized title "count"
table = pd.crosstab(index = carfeatures['cylinders'],columns = 'count')
table
```

```python
table.columns.name
```

```python
table.columns.name = 'column name'
table
```

```python
# "describe" computes the count, mean, std, min, 25% quantile, 50%, 75%, max
# automatically excludes variables with text values
# otherwise includes all numeric variables

carfeatures.describe()
```

```python
# display a scatter plot

plt.scatter(x = carfeatures['weight'], y = carfeatures['mpg'])
plt.show()
```
